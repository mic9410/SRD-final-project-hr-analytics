## Opis problemu
Dua firma o nazwie XYZ zatrudnia okoo 4000 pracownikw, jednak kadego roku okoo 15% z nich odchodzi z firmy i naley zastpi ich pul nowych ludzi dostpnych na rynku pracy. Zarzd uwaa, e ten poziom osabienia personelu (odejcie pracownikw na wasn rk lub dlatego, e zostali zwolnieni) jest niekorzystne dla firmy z nastpujcych powodw:

* Projekty byych pracownikw ulegaj opnieniu, co utrudnia dotrzymanie terminw, a w konsekwencji powoduje utrat reputacji wrd konsumentw i partnerw
* Naley utrzymywa sporych rozmiarw dziarekrutacji 
* Najczciej nowi pracownicy musz by przeszkoleni do pracy i / lub mie czas na aklimatyzacj w firmie

Dlatego zarzd zawar umow z firm analityczn HR, aby zrozumie, na jakich czynnikach powinni si skupi, aby ograniczy odejcia z firmy. Innymi sowy, chc wiedzie, jakie zmiany powinny zosta wprowadzone w miejscu pracy, aby wikszo pracownikw zostaa. Ponadto chc wiedzie, ktra z tych zmiennych jest najwaniejsza i naley od razu zaj si wnie tym czynnikiem.

## Cel analizy
Naley zamodelowa prawdopodobiestwo odejcia pracownika za pomoc regresji logistycznej. Uzyskane w ten sposb wyniki zostan wykorzystane przez kierownictwo do zrozumienia, jakie zmiany powinni wprowadzi w miejscu pracy, aby wikszo pracownikw pozostaa.
